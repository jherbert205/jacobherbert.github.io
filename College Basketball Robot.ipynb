This script downloads college basketball game data from sports-reference.com with the scores, ranks, and links to the box score, and team stats page from the date the games occurred for the entire 2018-2019 regular season.

```python
#import packages

from urllib.request import Request, urlopen
import datetime
import random
import time
```

```python
#download https://www.sports-reference.com/cbb/boxscores/index.cgi?month=11&day=9&year=2021

currentDate = datetime.date(2018, 11, 6)
endDate = datetime.date(2019, 3, 18)
delta = datetime.timedelta(days=1)

while currentDate<=endDate:

    timer = random.random()*2.5
    time.sleep
    
    year = str(currentDate)[0:4]
    month = str(currentDate)[5].strip('0')+str(currentDate)[6]
    day = str(currentDate)[8].strip('0')+str(currentDate)[9]
    

    url = f'https://www.sports-reference.com/cbb/boxscores/index.cgi?month={month}&day={day}&year={year}'
    req = Request(url)
    html = urlopen(req).read()
    
    fileout = f'DAYS{currentDate}.html'
    print ('Save to : ', fileout, '\n')

    #save files
    f = open(fileout,'w')
    f.write(html.decode('utf-8'))
    f.close()
    
    currentDate+=delta
    
```
